```python
name = input()
hour = int(input())
wage = int(input())
tax_rate1 = float(input())
tax_rate2 = float(input())

total_wage = hour * wage
minus = total_wage * (tax_rate1 + tax_rate2) 

print("="*15)
print("* 사원 이름 : %s"%name)
print("* 근무 시간 : %d시간"%hour)
print("* 시간당 급여 : %d원"%wage)
print("* 총 급여 : %d원"%(total_wage))
print("* 공제 : ")
print("원천징수세 (%.2f%%): %.1f원"%(tax_rate1, total_wage * tax_rate1))
print("주민세 (%.2f%%) : %.1f원"%(tax_rate2, total_wage * tax_rate2))
print("총 공제 : %.1f원"%minus)
print("공제 후 급여 : %d원"%(total_wage - minus))
```

```python
cMember = {'길동', '순신', '보검', '중근', '세종', '민주', '수지', '대건', '다산'}
late = {'길동', '보검', '수지'}
absent = {'민주', '수지', '보검'}
bonus = cMember - late - absent
night = late & absent

print("사원 명단 ", cMember)
print("지각자 명단 ", late)
print("결근자 명단 ", absent)
print("보너스 지급 대상자 ", bonus)
print("야근 대상자 ", night)
quit = set(input("퇴사자는? ").split())
new = input("신입사원? ")
cMember -= quit
cMember.add(new)
print("새 사원명단 ", cMember)
```
